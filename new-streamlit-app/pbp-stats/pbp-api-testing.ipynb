```python
import pandas as pd
import requests
```

```python
url = "https://api.pbpstats.com/get-totals/nba"
params = {
    "Season": "2025-26",
    "SeasonType": "Regular Season",
    "Type": "Team" # Use Opponent for opponent stats
}
response = requests.get(url, params=params)
response_json = response.json()
league_stats = response_json["single_row_table_data"]
team_stats = response_json["multi_row_table_data"]

league_averages = pd.DataFrame(league_stats, index=[0])
team_averages = pd.DataFrame(team_stats, index=range(len(team_stats)))

# Display all column names
pd.set_option('display.max_columns', None)
team_averages.columns.tolist()
```

```python
# Use params dictionary instead of building URL manually (cleaner and less error-prone)
url = "https://api.pbpstats.com/get-possessions/nba"
params = {
    "Season": "2025-26",  # Note: This might need to be "2024-25" if 2025-26 doesn't exist yet
    "SeasonType": "Regular Season",
    "TeamId": "1610612750",
    "OffDef": "Offense",
    "StartType": "OffMissedFG",
    "EventType": "FG2A"
}

poss_response = requests.get(url, params=params)

# Check response status and content before parsing
print(f"Status Code: {poss_response.status_code}")
print(f"Response Text (first 500 chars): {poss_response.text[:500]}")

# Only parse JSON if status is OK
if poss_response.status_code == 200:
    try:
        poss_response_json = poss_response.json()
        print(f"Response keys: {poss_response_json.keys() if isinstance(poss_response_json, dict) else 'Not a dict'}")
        
        if "multi_row_table_data" in poss_response_json:
            poss_df = pd.DataFrame(poss_response_json["multi_row_table_data"])
            print(f"\nDataFrame shape: {poss_df.shape}")
            poss_df.head()
        else:
            print("No 'multi_row_table_data' key in response")
            print(f"Available keys: {list(poss_response_json.keys())}")
    except requests.exceptions.JSONDecodeError as e:
        print(f"JSON decode error: {e}")
        print(f"Full response text: {poss_response.text}")
else:
    print(f"Error: HTTP {poss_response.status_code}")
    print(f"Response: {poss_response.text}")

```
