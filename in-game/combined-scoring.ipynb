```python
import nba_api.stats.endpoints
import requests
import json
import pandas as pd
import nba_api
```

```python
wolves_player_logs = nba_api.stats.endpoints.PlayerGameLogs(season_nullable='2025-26',team_id_nullable='1610612750').get_data_frames()[0]

wolves_player_logs
```

```python
wolves_player_logs[wolves_player_logs['PLAYER_NAME'] == 'Donte DiVincenzo']
```

```python
# Filter for Donte DiVincenzo and Jaden McDaniels
target_players = ['Donte DiVincenzo', 'Jaden McDaniels']
player_filtered = wolves_player_logs[wolves_player_logs['PLAYER_NAME'].isin(target_players)].copy()

# Group by game and sum their combined points
combined_pts = player_filtered.groupby('GAME_ID').agg({
    'PTS': 'sum',
    'GAME_DATE': 'first',
    'MATCHUP': 'first',
    'WL': 'first'  # Win/Loss column if available
}).reset_index()

combined_pts.columns = ['GAME_ID', 'COMBINED_PTS', 'GAME_DATE', 'MATCHUP', 'WL']

# Filter for games where combined PTS >= 30
games_30_plus = combined_pts[combined_pts['COMBINED_PTS'] >= 30].copy()

print(f"Games where Donte DiVincenzo and Jaden McDaniels combined for 30+ points: {len(games_30_plus)}")
print(f"\nGames breakdown:")
games_30_plus.sort_values('COMBINED_PTS', ascending=False).reset_index(drop=True)
```

```python
# Calculate win-loss record
# Check if WL column exists, if not parse from MATCHUP or other columns
if 'WL' in games_30_plus.columns and games_30_plus['WL'].notna().any():
    wins = (games_30_plus['WL'] == 'W').sum()
    losses = (games_30_plus['WL'] == 'L').sum()
else:
    # Try to parse from MATCHUP - if MIN is first, it's a home game
    # We'll need to check actual game results - let's check what columns are available
    print("WL column not found. Checking available columns...")
    print(games_30_plus.columns.tolist())
    # For now, we'll need to get win/loss from another source
    # Let's check if we can get it from the original dataframe
    game_results = wolves_player_logs.groupby('GAME_ID').agg({
        'WL': 'first'
    }).reset_index()
    
    games_30_plus = games_30_plus.merge(game_results, on='GAME_ID', how='left', suffixes=('', '_from_logs'))
    
    if 'WL_from_logs' in games_30_plus.columns:
        games_30_plus['WL'] = games_30_plus['WL_from_logs']
        wins = (games_30_plus['WL'] == 'W').sum()
        losses = (games_30_plus['WL'] == 'L').sum()
    else:
        wins = 0
        losses = 0
        print("Could not determine win/loss record from available data")

if wins + losses > 0:
    win_pct = wins / (wins + losses)
    print(f"\nTimberwolves Record when DiVincenzo + McDaniels combine for 30+ points:")
    print(f"Wins: {wins}")
    print(f"Losses: {losses}")
    print(f"Win Percentage: {win_pct:.3f} ({win_pct*100:.1f}%)")
    
    # Show the games
    print(f"\nGame Details:")
    display_cols = ['GAME_DATE', 'MATCHUP', 'COMBINED_PTS', 'WL']
    display_cols = [col for col in display_cols if col in games_30_plus.columns]
    print(games_30_plus[display_cols].sort_values('GAME_DATE', ascending=False))
else:
    print("No games found or unable to determine win/loss record")
```

```python
# Starting five: DiVincenzo, McDaniels, Anthony Edwards, Julius Randle, Rudy Gobert
starting_five = ['Donte DiVincenzo', 'Jaden McDaniels', 'Anthony Edwards', 'Julius Randle', 'Naz Reid', 'Rudy Gobert']
# starting_five = ['Naz Reid', 'Bones Hyland']

# Filter for starting five players
starting_five_logs = wolves_player_logs[wolves_player_logs['PLAYER_NAME'].isin(starting_five)].copy()

# Create a pivot table: games as rows, players as columns, PTS as values
pivot_pts = starting_five_logs.pivot_table(
    index='GAME_ID',
    columns='PLAYER_NAME',
    values='PTS',
    aggfunc='first'
).reset_index()

# Get game metadata
game_metadata = starting_five_logs.groupby('GAME_ID').agg({
    'GAME_DATE': 'first',
    'MATCHUP': 'first',
    'WL': 'first'
}).reset_index()

# Merge with pivot table
starting_five_games = pivot_pts.merge(game_metadata, on='GAME_ID', how='left')

# Check if all five players scored at least 10 points in each game
# First, ensure all five player columns exist (fill missing with 0)
for player in starting_five:
    if player not in starting_five_games.columns:
        starting_five_games[player] = 0

# Filter for games where all five scored at least 10 points
all_10_plus = starting_five_games[
    (starting_five_games['Naz Reid'] >= 10) &
    (starting_five_games['Jaden McDaniels'] >= 10) &
    (starting_five_games['Anthony Edwards'] >= 10) &
    (starting_five_games['Julius Randle'] >= 10) &
    (starting_five_games['Donte DiVincenzo'] >= 10) &
    (starting_five_games['Rudy Gobert'] >= 10)
].copy()

print(f"Games where all five starters scored 10+ points: {len(all_10_plus)}")
print(f"\nTotal games in dataset: {len(starting_five_games)}")
print(f"Percentage: {len(all_10_plus)/len(starting_five_games)*100:.1f}%")

# Display the games
if len(all_10_plus) > 0:
    display_cols = ['GAME_DATE', 'MATCHUP'] + starting_five + ['WL']
    display_cols = [col for col in display_cols if col in all_10_plus.columns]
    all_10_plus[display_cols].sort_values('GAME_DATE', ascending=False)
else:
    print("\nNo games found where all five starters scored 10+ points")
```

```python
# Calculate win-loss record for games where all five starters scored 10+ points
from IPython.display import display

if len(all_10_plus) > 0:
    # Ensure WL column exists
    if 'WL' not in all_10_plus.columns or all_10_plus['WL'].isna().all():
        # Get WL from original dataframe if missing
        game_results = wolves_player_logs.groupby('GAME_ID').agg({
            'WL': 'first'
        }).reset_index()
        all_10_plus = all_10_plus.merge(game_results, on='GAME_ID', how='left', suffixes=('', '_from_logs'))
        if 'WL_from_logs' in all_10_plus.columns:
            all_10_plus['WL'] = all_10_plus['WL_from_logs']
    
    wins = (all_10_plus['WL'] == 'W').sum()
    losses = (all_10_plus['WL'] == 'L').sum()
    
    if wins + losses > 0:
        win_pct = wins / (wins + losses)
        print(f"\nTimberwolves Record when all five starters score 10+ points:")
        print(f"Wins: {wins}")
        print(f"Losses: {losses}")
        print(f"Win Percentage: {win_pct:.3f} ({win_pct*100:.1f}%)")
        
        # Show detailed breakdown
        print(f"\nGame Details:")
        display_cols = ['GAME_DATE', 'MATCHUP', 
                       'Naz Reid', 'Donte DiVincenzo', 'Jaden McDaniels', 
                       'Anthony Edwards', 'Julius Randle', 'Rudy Gobert', 'WL']
        display_cols = [col for col in display_cols if col in all_10_plus.columns]
        
        # Calculate total points for each game
        all_10_plus_display = all_10_plus[display_cols].copy()
        player_cols = [col for col in starting_five if col in all_10_plus_display.columns]
        if player_cols:
            all_10_plus_display['TOTAL_PTS'] = all_10_plus_display[player_cols].sum(axis=1)
            display_cols = ['GAME_DATE', 'MATCHUP'] + player_cols + ['TOTAL_PTS', 'WL']
            display_cols = [col for col in display_cols if col in all_10_plus_display.columns]
        
        # Display the dataframe (using print like in Cell 4)
        result_df = all_10_plus_display[display_cols].sort_values('GAME_DATE', ascending=False)
        print(result_df)
    else:
        print("Could not determine win/loss record")
else:
    print("No games found where all five starters scored 10+ points")
```

```python
# Filter for Julius Randle games
randle_logs = wolves_player_logs[wolves_player_logs['PLAYER_NAME'] == 'Julius Randle'].copy()

# Check what column name is used for three-pointers made
# Common names: FG3M, FG3_M, FG3MADE, or similar
three_pt_col = None
for col in ['FG3M', 'FG3_M', 'FG3MADE', 'FG3A_MADE']:
    if col in randle_logs.columns:
        three_pt_col = col
        break

if three_pt_col is None:
    # List available columns to help identify the correct one
    print("Three-pointer column not found. Available columns:")
    print([col for col in randle_logs.columns if '3' in col.upper() or 'THREE' in col.upper()])
    # Try common alternative
    three_pt_col = 'FG3M' if 'FG3M' in randle_logs.columns else None

if three_pt_col:
    # Filter for games where Randle made at least 3 three-pointers
    randle_3plus_3pt = randle_logs[randle_logs[three_pt_col] >= 3].copy()
    
    print(f"Games where Julius Randle made 3+ three-pointers: {len(randle_3plus_3pt)}")
    print(f"\nTotal games played by Randle: {len(randle_logs)}")
    print(f"Percentage: {len(randle_3plus_3pt)/len(randle_logs)*100:.1f}%")
    
    # Calculate win-loss record
    if 'WL' not in randle_3plus_3pt.columns or randle_3plus_3pt['WL'].isna().all():
        # Get WL from original dataframe if missing
        game_results = wolves_player_logs.groupby('GAME_ID').agg({
            'WL': 'first'
        }).reset_index()
        randle_3plus_3pt = randle_3plus_3pt.merge(game_results, on='GAME_ID', how='left', suffixes=('', '_from_logs'))
        if 'WL_from_logs' in randle_3plus_3pt.columns:
            randle_3plus_3pt['WL'] = randle_3plus_3pt['WL_from_logs']
    
    wins = (randle_3plus_3pt['WL'] == 'W').sum()
    losses = (randle_3plus_3pt['WL'] == 'L').sum()
    
    if wins + losses > 0:
        win_pct = wins / (wins + losses)
        print(f"\nTimberwolves Record when Julius Randle makes 3+ three-pointers:")
        print(f"Wins: {wins}")
        print(f"Losses: {losses}")
        print(f"Win Percentage: {win_pct:.3f} ({win_pct*100:.1f}%)")
        
        # Show the games
        print(f"\nGame Details:")
        display_cols = ['GAME_DATE', 'MATCHUP', 'PTS', three_pt_col, 'WL']
        display_cols = [col for col in display_cols if col in randle_3plus_3pt.columns]
        print(randle_3plus_3pt[display_cols].sort_values('GAME_DATE', ascending=False))
    else:
        print("Could not determine win/loss record")
else:
    print("Could not find three-pointer column in the data")
```
