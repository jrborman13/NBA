PBPStats.com for Team Stats

```python
url = "https://api.pbpstats.com/get-totals/nba"
params = {
    "Season": "2024-25",
    "SeasonType": "Regular Season",
    "Type": "Team" # Use Opponent for opponent stats
}
response = requests.get(url, params=params)
response_json = response.json()
league_stats = response_json["single_row_table_data"]
team_stats = response_json["multi_row_table_data"]

league_averages = pd.DataFrame(league_stats, index=[0])
team_averages = pd.DataFrame(team_stats, index=range(len(team_stats)))

# Display all column names
pd.set_option('display.max_columns', None)
team_averages.columns.tolist()
```

PBPStats.com for Player Stats

```python
pbp_player_totals_url = "https://api.pbpstats.com/get-totals/nba"
pbp_player_totals_params = {
    "Season": "2025-26",
    "SeasonType": "Regular Season",
    "Type": "Player" # Use Opponent for opponent stats
}
pbp_player_totals_response = requests.get(pbp_player_totals_url, params=pbp_player_totals_params)
pbp_player_totals_response_json = pbp_player_totals_response.json()
player_stats = pbp_player_totals_response_json["multi_row_table_data"]

player_averages = pd.DataFrame(player_stats, index=range(len(player_stats)))

player_averages.columns.tolist()
```

```python
player_averages['FGA'] = player_averages['FG2A'] + player_averages['FG3A']
player_averages['FGM'] = player_averages['FG2M'] + player_averages['FG3M']

player_averages[player_averages['FGA'] >= 125].sort_values(by='TsPct', ascending=False).reset_index(drop=True).head(10)[[
    'Name', 'RowId', 'TeamAbbreviation', 'FGM', 'FGA', 'TsPct'
]]
```
