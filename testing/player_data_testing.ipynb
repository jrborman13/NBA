```python
current_season = '2024-25'

game_logs_ex = nba_api.stats.endpoints.PlayerGameLogs(
    league_id_nullable='00',
    season_nullable=current_season
).get_data_frames()[0]
```

```python
game_logs_ex
```

```python
game_logs_ex.columns
```

```python
grouped_log = game_logs_ex.loc[game_logs_ex['PLAYER_NAME'].isin(['Donte DiVincenzo', 'Naz Reid'])].groupby(['GAME_DATE','MATCHUP', 'GAME_ID',  'WL']).agg({'PTS': 'sum'}).reset_index().sort_values(by='GAME_DATE', ascending=True)

grouped_log_filtered = grouped_log[grouped_log['PTS'] <= 13]

grouped_log_filtered.groupby(['WL']).agg({'PTS': 'count'}).reset_index()
```

```python
ant_ju_logs = game_logs_ex[
    (game_logs_ex['PLAYER_NAME'].isin(['Anthony Edwards', 'Julius Randle'])) &   
    (game_logs_ex['GAME_ID'].isin(grouped_log_filtered['GAME_ID']))].sort_values(by='GAME_DATE', ascending=True).groupby(['GAME_DATE','MATCHUP', 'GAME_ID',  'WL']).agg({'PTS': 'sum'}).reset_index().sort_values(by='GAME_DATE', ascending=True)


ant_ju_logs.groupby(['WL']).agg({'PTS': 'count'}).reset_index()
```

```python

ddv = game_logs_ex.loc[game_logs_ex['PLAYER_NAME'] == 'Donte DiVincenzo'].sort_values(by='GAME_DATE')

ddv = ddv.assign(game_num=range(1, len(ddv) + 1))

ddv
```

```python
ddv.columns
```

```python
# Create a DataFrame for plotting
data = pd.DataFrame({
    'x': range(10),
    'y1': [i * 1.5 for i in range(10)],
    'y2': [i * 2.5 for i in range(10)],
    'y3': [i * 3.5 for i in range(10)]
})

# Create individual line charts
# Create the line chart
chart = alt.Chart(ddv).mark_line().encode(
    x='game_num:Q',  # Quddvitative (numerical) x-axis
    y='PTS:Q'        # Quddvitative (numerical) y-axis
).properties(
    title="Points per Game"
)

# Display the chart
chart
```

```python
# Calculate 5-game moving average
ddv['PTS_MA'] = ddv['PTS'].rolling(window=5, min_periods=1).mean()

# Create the base chart for actual points
pts_chart = alt.Chart(ddv).mark_line(color='#f76517').encode(
    x='game_num:Q',
    y='PTS:Q',
    tooltip=['game_num', 'PTS']
).properties(
    title="DDV PPG with 5-Game Moving Average"
)

# Create the moving average line
ma_chart = alt.Chart(ddv).mark_line(color='#175aaa').encode(
    x='game_num:Q',
    y='PTS_MA:Q',
    tooltip=['game_num', 'PTS_MA']
)

# Combine both charts
final_chart = pts_chart + ma_chart
final_chart
```

```python
player_splits = nba_api.stats.endpoints.PlayerDashboardByGeneralSplits(
    season=current_season,
    measure_type_detailed = 'Base',
    per_mode_detailed = 'PerGame',
    player_id=ddv_id,
    season_type_playoffs = 'Regular Season'
)

player_splits
```

```python
ddv_starting_splits = player_splits.get_data_frames()[5][[
    'GROUP_VALUE', 'GP', 'W', 'L', 'W_PCT', 
    'MIN','PTS', 'FGM', 'FGA', 'FG_PCT', 
    'FG3M', 'FG3A', 'FG3_PCT', 
    'FTM', 'FTA', 'FT_PCT', 
    'OREB', 'DREB', 'REB', 'AST', 'TOV', 
    'STL', 'BLK',
    'PLUS_MINUS'
]]
```

```python
# Format different column types appropriately
ddv_starting_splits.style.format({
    # Percentage columns
    'FG_PCT': '{:.1%}',
    'FG3_PCT': '{:.1%}',
    'FT_PCT': '{:.1%}',
    'W_PCT': '{:.1%}',
    
    # Integer columns (no decimals)
    'GP': '{:.1f}',
    'W': '{:.1f}',
    'L': '{:.1f}',
    'MIN': '{:.1f}',
    'PTS': '{:.1f}',
    'FGM': '{:.1f}',
    'FGA': '{:.1f}',
    'FG3M': '{:.1f}',
    'FG3A': '{:.1f}',
    'FTM': '{:.1f}',
    'FTA': '{:.1f}',
    'OREB': '{:.1f}',
    'DREB': '{:.1f}',
    'REB': '{:.1f}',
    'AST': '{:.1f}',
    'TOV': '{:.1f}',
    'STL': '{:.1f}',
    'BLK': '{:.1f}',
    'PF': '{:.1f}',
    'PLUS_MINUS': '{:.1f}'
}).set_table_styles([
    {'selector': 'th', 'props': [('background-color', '#f2f2f2')]},
    {'selector': 'td', 'props': [('text-align', 'center')]}
])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))
ax.axis('tight')
ax.axis('off')

# Create table
table = ax.table(cellText=ddv_starting_splits.values,
                colLabels=ddv_starting_splits.columns,
                cellLoc='center',
                loc='center')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.5)

# Color the header
for i in range(len(ddv_starting_splits.columns)):
    table[(0, i)].set_facecolor('#40466e')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Save as image
plt.title('Donte DiVincenzo Starting Splits', fontsize=16, fontweight='bold', pad=20)
plt.savefig('/Users/jackborman/Downloads/ddv_starting_splits.png', dpi=300, bbox_inches='tight')
plt.show()
```
