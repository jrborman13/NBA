```python
import nba_api
import nba_api.stats.endpoints
import pandas as pd
import requests
```

```python
paint_scoring = nba_api.stats.endpoints.LeagueDashPlayerStats(
    measure_type_detailed_defense = 'Misc',
    per_mode_detailed = 'PerGame', 
    season = '2025-26',
    season_type_all_star = 'Regular Season'
)

paint_scoring_df = paint_scoring.get_data_frames()[0]

paint_scoring_df.sort_values(by='PTS_PAINT', ascending=False, inplace=True)

paint_scoring_df
```
